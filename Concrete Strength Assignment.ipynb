# Load Data:
[Click Here to Download DataSet](https://github.com/ramsha275/ML_Datasets/blob/main/compresive_strength_concrete.csv)

```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("archive/compresive_strength_concrete.csv", header=None)
dataset
```

```python
# droping first row of the dataset which includes the headers names
dataset.drop(index=0, inplace=True)
dataset
```

```python
# converting all values to numeric from string
dataset = dataset.apply(pd.to_numeric)
```

```python
dataset.describe()
```

# Checking for NaN values 

```python
dataset.isnull().sum()
```

```python
dataset.isna().sum()
```

# Spliting the data

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
X_train, X_test = train_test_split(dataset, train_size = 0.5, test_size = 0.5, random_state = 100)

X_test, X_val = train_test_split(dataset, train_size = 0.3, test_size = 0.2, random_state = 100)
```

```python
# saving the output column i.e Concrete Comprehensive Strength in a separate dataframe
Y_train = X_train[8]
# now droping the price column from the input dataset
X_train.drop(columns=8, inplace=True)
```

```python
Y_test = X_test[8]
# now droping the price column from the input dataset
X_test.drop(columns=8, inplace=True)
```

```python
Y_val = X_val[8]
# now droping the price column from the input dataset
X_val.drop(columns=8, inplace=True)
```

# Normalizing data

```python
xtrain_mean = X_train.mean(axis=0)
xtrain_std = X_train.std(axis=0)
X_train -= xtrain_mean
X_train /= xtrain_std
X_train
```

```python
ytrain_mean = Y_train.mean(axis=0)
ytrain_std = Y_train.std(axis=0)
Y_train -= ytrain_mean
Y_train /= ytrain_std
```

```python
xtest_mean = X_test.mean(axis=0)
xtest_std = X_test.std(axis=0)
X_test -= xtest_mean
X_test /= xtest_std
```

```python
ytest_mean = Y_test.mean(axis=0)
ytest_std = Y_test.std(axis=0)
Y_test -= ytest_mean
Y_test /= ytest_std
```

```python
xval_mean = X_val.mean(axis=0)
xval_std = X_val.std(axis=0)
X_val -= xval_mean
X_val /= xval_std
```

```python
yval_mean = Y_val.mean(axis=0)
yval_std = Y_val.std(axis=0)
Y_val -= yval_mean
Y_val /= yval_std
```

```python
y_test = Y_test.values
x_test = X_test.values
```

# Building the model

```python
# building the model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import regularizers

model = Sequential() # Initialising the ANN

model.add(Dense(units = 10, activation = 'relu', kernel_regularizer=regularizers.l2(0.02), input_shape = (X_train.shape[1],)))
model.add(Dense(units = 8, activation = 'relu', kernel_regularizer=regularizers.l2(0.02)))
model.add(Dense(units = 6, activation = 'relu', kernel_regularizer=regularizers.l2(0.02)))
model.add(Dense(units = 1))
```

```python
model.compile(optimizer = 'adam', loss = 'mse', metrics='mae')
```

```python
# training the data
history = model.fit(X_train, Y_train, batch_size = 20, epochs = 100, validation_data=(X_val, Y_val))
```

```python
history_dict = history.history
```

```python
history_dict.keys()
```

```python
import matplotlib.pyplot as plt
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
```

```python
# plot a graph of loss and val loss
epochs = range(100)
plt.plot(epochs,loss_values,'bo',label='Training Loss')
plt.plot(epochs,val_loss_values,'b',label='Validation Loss')
plt.title('Training and Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
mae = history_dict['mae']
val_mae = history_dict['val_mae']
```

```python
# plot a graph between mae and val_mae
epochs = range(100)
plt.plot(epochs,mae,'bo',label='Mean Absolute Error')
plt.plot(epochs,val_mae,'b',label='Validation Mean Absolute Error')
plt.title('Training and Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# Evaluating the model

```python
model.evaluate(X_test,Y_test)
```

# Prediction

```python
# predicting data from X_test
ynew = model.predict(X_test)
```
